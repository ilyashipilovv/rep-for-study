```python
avg_age_medals = og_data.copy()
```

```python
avg_age_medals = avg_age_medals.dropna()
avg_age_medals = avg_age_medals.groupby('Medal')['Age'].mean()
avg_age_medals = dict(avg_age_medals.round(1))
medals = ['Bronze', 'Silver', 'Gold']
for k in medals:
    print(f'Средний возраст спортсменов, получивших {k} равен {avg_age_medals[k]} лет')
```

```python
nan_counts = og_data.isna().sum()
nan_counts
```

```python
for k, v in nan_counts.items():
    print(f'процент NaN значений в столбце {k} равен {round(v / 70000 * 100, 1)}%')
```

```python
medals_rank = og_data.copy().dropna()
```

```python
gsb = pd.get_dummies(medals_rank.Medal)
```

```python
medals_rank = pd.concat([medals_rank, gsb], axis=1)
cols = ['Bronze', 'Gold', 'Silver']
medals_rank[cols] = medals_rank[cols].astype(int)
```

```python
medals_rank = medals_rank.groupby('Team')[['Gold', 'Silver', 'Bronze']].sum()
```

```python
sorted_medals_rank = medals_rank.sort_values(by=['Gold', 'Silver', 'Bronze'], ascending=False)
sorted_medals_rank.head(20)
```

```python
sorted_medals_rank['Total'] = sorted_medals_rank['Gold'] + sorted_medals_rank['Silver'] + sorted_medals_rank['Bronze']
```

```python
sorted_medals_rank.head(30)
```

```python
sportsmen_medals = og_data.copy().dropna()
```

```python
gsb = pd.get_dummies(sportsmen_medals.Medal)
```

```python
sportsmen_medals = pd.concat([sportsmen_medals, gsb], axis=1)
cols = ['Bronze', 'Gold', 'Silver']
sportsmen_medals[cols] = sportsmen_medals[cols].astype(int)
```

```python
sportsmen_medals
```

```python
sportsmen_medals = sportsmen_medals.groupby('Name')[['Gold', 'Silver', 'Bronze']].sum()
```

```python
sportsmen_medals 
```

```python
sort_sportsmen = sportsmen_medals.sort_values(by=['Gold', 'Silver', 'Bronze'], ascending=False)
```
