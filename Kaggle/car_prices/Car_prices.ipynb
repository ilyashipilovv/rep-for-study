```python
for brand in car_prices_ds['Brand'].unique():
    car_brand = car_prices_ds[car_prices_ds.Brand == f'{brand}']
    low_high = [car_brand.Price.min(), car_brand.Price.median(), car_brand.Price.max()]
    print(f'Цена самого дешевого автомобиля марки {brand} = {low_high[0]}')
    print(f'Медианная цена автомобилей марки {brand} = {low_high[1]}')
    print(f'Цена самого дорогого автомобиля марки {brand} = {low_high[2]}')
    print('--------------------------------------------------------')
```
